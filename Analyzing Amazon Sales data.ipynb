```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Convert date columns to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
```

```python
# Check for missing values
missing_values = df.isnull().sum()
```

```python
# Summary statistics
summary_statistics = df.describe()
missing_values, summary_statistics
```

```python
df.info()
```

```python
# Plot total revenue over time
plt.figure(figsize=(14, 7))
df.groupby('Order Date')['Total Revenue'].sum().plot()
plt.title('Total Revenue Over Time')
plt.xlabel('Order Date')
plt.ylabel('Total Revenue')
plt.grid(True)
plt.show()
```

```python
# Summarize sales and profit by item type
item_sales_profit = df.groupby('Item Type')[['Total Revenue', 'Total Profit']].sum().sort_values(by='Total Revenue', ascending=False)
```

```python
# Plot sales and profit by item type
plt.figure(figsize=(14, 7))
sns.barplot(x=item_sales_profit.index, y=item_sales_profit['Total Revenue'], palette='viridis')
plt.title('Total Revenue by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Revenue')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 7))
sns.barplot(x=item_sales_profit.index, y=item_sales_profit['Total Profit'], palette='viridis')
plt.title('Total Profit by Item Type')
plt.xlabel('Item Type')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)
plt.show()
```
